```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/homeprices.csv")
df
```

```python
type(df['price'])
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price($)')
plt.title("HOUSEPREDICTION")
plt.scatter(df['area'],df['price'], color='red', marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df['price'])
```

```python
reg.predict(np.array([[3300]])) #accepting 2d array
```

```python
reg.coef_ # value for m(slope), in y = mx + b
```

```python
reg.intercept_ #value for intercept b
```

```python
# y = m*x + b
135.78767123*3300+180616.43835616432
```

```python
reg.predict(df[['area']])
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price($)')
plt.title("HOUSEPREDICTION")
plt.scatter(df['area'],df['price'], color='red', marker = '+')
plt.plot(df['area'],reg.predict(df[['area']]), color='blue')
```
